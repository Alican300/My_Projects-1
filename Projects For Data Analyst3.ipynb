```python
# The Best Selling Books
```

```python
df["Selling_receive"]=df["Reviews"]*df["Price"]
```

```python
df.head(5)
```

```python
df_TheBestSelling=df.groupby(["Author","Name"])["Selling_receive"].sum().reset_index()
```

```python
df_TheBestSelling.sort_values(by="Selling_receive",ascending=False).reset_index()
```

```python
#The Top 10 Selling
Top_10=df_TheBestSelling.sort_values(by="Selling_receive",ascending=False).head(10).reset_index()
```

```python
Top_10
```

```python
#Average User Rating for Authors
df.groupby("Author")["User Rating"].mean()
```

```python
Authors=df.groupby("Author")["User Rating"].mean().reset_index()
```

```python
Authors
```

```python
#The Three Most Liked Authors
df_most_liked_authors=df.groupby("Author")["User Rating"].mean()
df_most_liked_authors.sort_values(ascending=False).reset_index().head(3)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=Top_10,x="Name",y="Selling_receive",hue="Author")
plt.title("Best Sellers")
plt.xlabel("Authors")
plt.ylabel("Selling_receive")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=Top_5_books,x="Author",y="Reviews",hue="Year")
plt.title("Best Authors")
plt.xlabel("Authors")
plt.ylabel("Total Reviews")
plt.tight_layout()
```
