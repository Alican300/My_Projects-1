```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("sales_data_sample.csv",encoding="latin1")
```

```python
df
```

```python
df.dropna(inplace=True)
df["Order_Date"]=pd.to_datetime(df["ORDERDATE"])
df["Year"]=df["Order_Date"].dt.year
df["Month"]=df["Order_Date"].dt.month
```

```python
df
```

```python
total_revenue=df["SALES"].sum()
print("total sales",total_revenue)
```

```python
mounthly_group=df.groupby(["Year","Month"])
```

```python
mounthly_group.count()
```

```python
df["Year"].unique()
```

```python
mounthly_sales=mounthly_group["SALES"].sum()
```

```python
mounthly_sales
```

```python
top_products=df.groupby("PRODUCTCODE")["ORDERNUMBER"].sum().sort_values(ascending=False).head(10)
```

```python
top_products
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df,x="Month",y="SALES",hue="Year",marker="o")
plt.title("Mounthly Sales Trend")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.show()
            
```
