```python
X=df.drop("Outcome",axis=1)
y=df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=15)
```

```python
columns_to_fill=["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]
```

```python
medians={}
for col in columns_to_fill:
    median_value=X_train[X_train[col]!=0][col].median()
    medians[col]=median_value
    X_train[col]=X_train[col].replace(0,median_value)
    
for col in columns_to_fill:
    X_test[col]=X_test[col].replace(0,medians[col])
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
ada=AdaBoostClassifier()
ada.fit(X_train,y_train)
y_pred=ada.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
def calculate_model_metrcis(true,predict):
    accuracyScore=accuracy_score(true,predict)
    classificationReport=classification_report(true,predict)
    confusionMatrix=confusion_matrix(true,predict)
    return accuracyScore,classificationReport,confusionMatrix
```

```python
models={
    "RandomForestClassifier":RandomForestClassifier(),
    "DecisionTreeClassifier":DecisionTreeClassifier(),
    "LogisticRegression":LogisticRegression(),
    "KNeighborsClassifier":KNeighborsClassifier()
}
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)
    
    y_train_pred=model.predict(X_train)
    y_test_pred=model.predict(X_test)
    
    accuracyScore_train,classificationReport_train,confusionMatrix_train=calculate_model_metrcis(y_train,y_train_pred)
    accuracyScore_test,classificationReport_test,confusionMatrix_test=calculate_model_metrcis(y_test,y_test_pred)
    
    print(list(models.values())[i])
    print("-------------------")
    
    print("Evaluation for Train Set")
    print("Accuracy_Score",accuracyScore_train)
    print("Classification_Report",classificationReport_train)
    print("Confusion_matrix",confusionMatrix_train)
    
    print("-------------------")
    
    print("Evaluation for Test Set")
    print("Accuracy_Score",accuracyScore_test)
    print("Classification_Report",classificationReport_test)
    print("Confusion_matrix",confusionMatrix_test)
    
    print("-------------------")
    print("\n")
```

```python
Kn_params={
    "n_neighbors":[1,5,10,15,20,25,50,100],
    
}
Logistic_params={
    "penalty":["l1","l2","elasticnet"],
    "C":[0.001,0.01,0.1,1],
    "solver":["lbfgs","liblinear","newton-cg","newton-cholesky","sag","saga"]
}    
```

```python
grid_cv_model=[
    ("Kn",KNeighborsClassifier(),Kn_params),
    ("Log",LogisticRegression(),Logistic_params)
]
```

```python
for name,model,param in grid_cv_model:
    grid_new=GridSearchCV(estimator=model,param_grid=param,cv=5,n_jobs=-1,verbose=1)
    grid_new.fit(X_train,y_train)
    print("best params of",name,grid_new.best_params_)
```
