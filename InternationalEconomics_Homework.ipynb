```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
df_USA = pd.read_excel("USA-Partner.xlsx", sheet_name="Partner")
df_USA.head(10)
```

```python
df_TUR = pd.read_excel("TUR-Partner.xlsx", sheet_name="Partner")
df_GER=pd.read_excel("GER-Partner.xlsx", sheet_name="Partner")
df_ARG=pd.read_excel("ARG-Partner.xlsx", sheet_name="Partner")
df_BRA=pd.read_excel("BRA-Partner.xlsx", sheet_name="Partner")
df_RUS=pd.read_excel("RUS-Partner.xlsx", sheet_name="Partner")
df_MEX=pd.read_excel("MEX-Partner.xlsx", sheet_name="Partner")
df_FRA=pd.read_excel("FRA-Partner.xlsx", sheet_name="Partner")
df_ENG=pd.read_excel("ENG-Partner.xlsx", sheet_name="Partner")
df_JAP=pd.read_excel("JAP-Partner.xlsx", sheet_name="Partner")
```

```python
df_USA[["Reporter Name","Partner Name","Year","Trade Flow","Import (US$ Thousand)"]].sort_values(by="Import (US$ Thousand)",ascending=False).reset_index(drop=True).head(10)
```

```python
df_USA=df_USA[["Reporter Name","Partner Name","Year","Trade Flow","Import (US$ Thousand)"]].sort_values(by="Import (US$ Thousand)",ascending=False).reset_index(drop=True).head(10)
```

```python
country_list=[df_USA,df_GER,df_JAP]
country_names=["United States","Germany","Japan"]
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(15,5))
for i,col, in enumerate(country_list):
    sns.barplot(data=col,x="Import (US$ Thousand)",y="Partner Name",ax=ax[i])
    plt.suptitle("Import of Countries",fontsize=13)
    ax[i].set_title(country_names[i])
    ax[i].set_ylabel("Countries")
    ax[i].set_xlabel("Ä°mport(US$ Thousand)")

plt.tight_layout()
plt.show()
```
