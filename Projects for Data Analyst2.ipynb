```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Sample - Superstore.csv",encoding="latin1")
```

```python
total_sales=df["Sales"].sum()
print("Total sales",total_sales)
total_profit=df["Profit"].sum()
print("Total profit",total_profit)
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Month"]=df["Order Date"].dt.month
df["Year"]=df["Order Date"].dt.year
```

```python
df[["Year","Month"]]
```

```python
groupby_month=df.groupby("Month")["Sales"].sum()
```

```python
groupby_month.count()
```

```python
plt.figure(figsize=(7,4))
sns.lineplot(data=groupby_month,x=groupby_month.index,y=groupby_month.values)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales by Month")
plt.show()
```

```python
groupby_month.index
```

```python
top_cities=df.groupby("City")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
top_cities
```

```python
top_cities.plot(kind="bar",figsize=(10,5),title="The Most Selling Cities")
plt.ylabel("Sales")
plt.show()
```

```python
category_sales=df.groupby("Category")["Sales"].sum()
```

```python
category_sales.plot(kind="pie",autopct='%1.1f%%',title="Sales by Category")
plt.ylabel("")
plt.show()
```

```python
top_subcategories=df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
top_subcategories.plot(kind="bar",x=top_subcategories.index,y=top_subcategories.values)
plt.title("Top Subcategories")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```
